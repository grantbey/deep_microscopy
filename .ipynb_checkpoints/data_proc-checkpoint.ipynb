### Weighted masks

This section deals with making weighted masks to force the network to learn the small boundaries between nuclei. It is based on the details of the original U-Net paper (found [here](https://arxiv.org/pdf/1505.04597.pdf)).

```python
# Load the images
img = cv2.imread('./data/raw/x_original.tif')[:,:,[2,1,0]]
msk = cv2.imread('./data/raw/nuclei_mask.tif',cv2.CV_32SC1)
edt = cv2.imread('./data/raw/nuclei_maskEDT.tif',cv2.IMREAD_ANYDEPTH)

# Rescale the EDT image such close pixels are higher in value but leave 0-value pixels as-is
edt[edt!=0] = 255/edt[edt!=0]
```

```python
img = img[200:500,200:500]
msk = msk[200:500,200:500]
edt = edt[200:500,200:500]
```

```python
# find the nucleus contours in the mask
nuclei_contours = cv2.findContours(msk.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)[1]
print('Found {} nuclei'.format(len(nuclei_contours)))
```

```python
# Make list of black pixels to check
blacks = [[j,i] for j in range(edt.shape[1]) for i in range(edt.shape[0]) if edt[i,j] != 0 and edt[i,j] > 255/3]
print('{} pixels to measure'.format(len(blacks)))
```

```python
#%%timeit -n 1

# Calculate the weights!
# Uses pointPolygonTest

# sigma set as recommended in the paper
sigma = 5 

# Initialize empty mask
weighted_msk = np.zeros_like(msk)

# Complicated list comprehension which measures the distance for each pixel for each nucleus
# Then the list of distances for that pixel are sorted, and the first two are selected
# The pixel's x,y coordinates are also returned
# Note: output is a list of lists, with each element being [[dist1,dist2],row,column]
measurements = [[sorted([np.abs(cv2.pointPolygonTest(nucleus,(pixel[0],pixel[1]),True)) for nucleus in nuclei_contours])[:2],pixel[0],pixel[1]] for pixel in blacks]

# For each pixel, calculate the weight and fill the value into the empty mask
for i,pixel in enumerate(measurements):
    weighted_msk[pixel[2],pixel[1]] = 10 * np.exp( -1 * (((np.abs(pixel[0][0])+np.abs(pixel[0][1]))**2) / (2*(sigma**2))) )
    if i % 10000 == 0:
        print('Completed pixel {}'.format(i+1))
```

```python
# Make training images and masks

# Specify the image size variables and the h5py file name
img_h,img_w = 3040,4000
filename = 'mosaic_1'

# Check if h5py file exists, if not create it
# Note the directory must exist otherwise this will fail
if not os.path.isfile('./data/training-data/{}.hdf5'.format(filename)):
    h5py.File('./data/training-data/{}.hdf5'.format(filename),'w')
    print('Writing the file...')

# Open the h5py file
# Note: the "with" here means that the file is opened, used, and then closed automatically
with h5py.File('./data/training-data/{}.hdf5'.format(filename),'r+') as f:
    
    # Create the x,y datasets
    # Each dataset is an array with four dimensions:
    # (images,x_dimension, y_dimension, channels)
    # I specify the size of the dimensions here using the shape attribute
    # Neural networks usually require float32 data types. The y masks are stored as int8 to preserve space (1s and 0s)
    # The network will convert these to float32 automatically. Given the small size of the dataset this is probably unnecessary
    f.require_dataset('x',shape=(100,304,400,3),dtype='float32')
    f.require_dataset('y_nuclei',shape=(100,304,400,1),dtype='int8')
    f.require_dataset('y_nuclei_weighted',shape=(100,304,400,1),dtype='float32')
    f.require_dataset('y_infected',shape=(100,304,400,1),dtype='int8')

    # Read the image files into arrays using cv2s imread method
    # Normalize to [0,1] by dividing by 255
    img = cv2.imread('./data/raw/x_original.tif')[:,:,[2,1,0]]/255
    msk_infected = cv2.imread('./data/raw/infected_mask.tif')[:,:,0:1]/255
    msk_nuclei = cv2.imread('./data/raw/nuclei_mask.tif')[:,:,0:1]/255
    #msk_nuclei_EDT = cv2.imread('./data/raw/nuclei_maskEDT.tif')[:,:,0:1]/255
    msk_nuclei_weighted = cv2.imread('./data/raw/nuclei_msk_weighted.tif')[:,:,0:1]
    
    # Resize everything to 3000,4000 for simplicity's sake
    # Note: original size was 3000,3881 so mild horizontal stretch
    for item in [img,msk_infected,msk_nuclei,msk_nuclei_weighted]:
        item = skimage.transform.resize(item,(img_h,img_w),order=3)
                
    # Split images into tiles and save each one to the datasets
    for j in range(10):
        for k in range(10):
            # Create an image id, which is just an E(0,100]
            img_idx = 10*j+k
            # The row/col indices are used to slice the images below
            row_idx = j*100
            col_idx = k*100
            # Here we slice the arrays, assiging the result to it's specific image id in the datasets
            f['x'][img_idx,...] = img[row_idx:row_idx+304,col_idx:col_idx+400,:]
            f['y_nuclei'][img_idx,...] = msk_nuclei[row_idx:row_idx+304,col_idx:col_idx+400,:]
            f['y_nuclei_weighted'][img_idx,...] = msk_nuclei_weighted[row_idx:row_idx+304,col_idx:col_idx+400,:]
            f['y_infected'][img_idx,...] = msk_infected[row_idx:row_idx+304,col_idx:col_idx+400,:]
            

# Send push notification that the image processing is done
push('The new image arrays are done')
```
